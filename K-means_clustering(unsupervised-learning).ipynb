```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Iris dataset

iris = load_iris()

X = pd.DataFrame(iris.data, columns=iris.feature_names)

y = iris.target  # Not used in clustering since it's unsupervised
```

```python
# Display type and shape of features

print(f"Type of X: {type(X)}")

print(f"Shape of X: {X.shape}")
```

```python
# Define number of clusters
num_clusters = 3
# Initialize KMeans with fixed random state for reproducibility
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
# Fit the model to data
kmeans.fit(X)
# Get cluster centers (centroids)
centroids = kmeans.cluster_centers_
# Predict cluster for each data point
clusters = kmeans.predict(X)
```

```python
# Add cluster labels to the DataFrame
X['Cluster'] = clusters
# Plot clusters with seaborn pairplot
sns.pairplot(X, hue='Cluster', diag_kind='kde', palette='Set1')
# Mark centroids on the plot (example on sepal length vs sepal width)
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='black', marker='X', label='Centroids')
plt.legend()
plt.suptitle('K-Means Clustering on Iris Dataset')
plt.show()
```
